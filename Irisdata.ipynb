# Lets build a manual model

```python
df_train["petal length (cm)"].values
```

```python
def simple_predict(petal_length): #predicting iris species on basis of petal length
    if petal_length < 2.5:
        return 0
    if petal_length < 4.0:
        return 1
    else:
        return 2
```

```python
def plot_pred(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2,2, figsize=(10,10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y = y_axis_feature, hue = "prediction_label", data =df_predictions, ax = axs[0])
    sns.scatterplot(x=x_axis_feature, y = y_axis_feature, hue = "target_name", data =df_predictions, ax = axs[1])
    sns.scatterplot(x=x_axis_feature, y = y_axis_feature, hue = "correct_prediction", data =df_predictions, ax = axs[2])
    axs[3].s

    plt.show()
```

```python
fig, axs =  plt.subplots(2,2, figsize=(10,10))
```

```python
plot_pred(df_predictions, "petal length (cm)", "petal width (cm)")
```
